# Linear Regression with single variable  |  Linear Regression with Multiple  variable

# Linear Regression with single variable

```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
var = pd.DataFrame({"age":[25,30,35,40,45],"premium":[18000,32000,40000,47000,55000]})
var
```

```python
sns.lmplot(x='age',y='premium',data=var)
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(var[['age']],var[['premium']])
```

```python
reg.predict([[21]])
```

```python
reg.predict([[50]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

# Linear Regression with Multiple  variable

```python
import pandas as pd
from sklearn import linear_model
```

```python
var = pd.DataFrame({"age":[25,30,35,40,45],'height':[162.56,172.72,167.64,'NaN',157.48],'weight':[70,95,78,110,85],"premium":[18000,32000,40000,47000,55000]})
var
```

```python
var['height'] = pd.to_numeric(var['height'], errors='coerce')
mean_height= var['height'].mean()
```

```python
var.height
```

```python
var.height= var.height.fillna(mean_height)
```

```python
var
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(var[['age','height','weight']],var[['premium']])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[27,167.56,60]])
```
